```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
d0="Geeks for geeks"
d1="Geeks"
d2="P2j"
string=[d0,d1,d2]
```

```python
tfidf=TfidfVectorizer()
result=tfidf.fit_transform(string)
```

```python
print("\n idf values:")
for ele1,ele2 in zip(tfidf.get_feature_names_out(),tfidf.idf_):
    print(ele1,":",ele2)
```

```python
print("\n Word indexes:")
print(tfidf.vocabulary_)
print("\n tf-idf value")
print(result)
print("\n tf-idf in matrix form:")
print(result.toarray())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Assuming 'result' is your TF-IDF matrix (from TfidfVectorizer)
# and 'tfidf' is your fitted TfidfVectorizer object
tfidf_dense = result.toarray()

# Create a DataFrame with feature names (words)
df = pd.DataFrame(tfidf_dense, columns=tfidf.get_feature_names_out())

# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df, cmap="YlGnBu", cbar=True)  # removed annot=True (too cluttered for TF-IDF)
plt.title("TF-IDF Heatmap")
plt.xlabel("Words")
plt.ylabel("Documents")
plt.show()
```
