```python
from gensim.models	import	Word2Vec
from gensim.utils	import	simple_preprocess
```

```python
corpus	=	[
				"Machine	learning	is	fun	and	powerful",
				"Deep	learning	is	a	subset	of	machine	learning",
				"Natural	language	processing	helps	computers	understand	human	language",
				"Artificial	intelligence	and	machine	learning	are	closely	related",
				"Python	is	a	great	programming	language	for	data	science"
 ]
```

```python
tokenized_corpus	=	[simple_preprocess(sentence)	for	sentence	in	corpus]
```

```python
model	=	Word2Vec(
				sentences=tokenized_corpus,
				vector_size=100,				#	Dimensionality	of	word	embeddings
				window=5,											#	Context	window	size
				min_count=1,								#	Minimum	frequency	for	words
				sg=1,															#	Use	skip-gram;	set	to	0	for	CBOW
				epochs=100										#	Number	of	training	iterations
 )
```

```python
print("\nMost	similar	to	'learning':")
print(model.wv.most_similar('learning',	topn=5))
```

```python
similarity_score	=	model.wv.similarity('machine',	'learning')
print(f"\nSimilarity	between	'machine'	and	'learning':	{similarity_score:.4f}")
```
