```python
import	re
```

```python
patterns={
				r'hi|hello|hey':	"Hello!	How	can	I	help	you	today?",
				r'how	are	you':	"I'm	a	bot,	but	I'm	doing	great!	How	about	you?",
				r'what	is	your	name':	"I'm	a	simple	rule-based	chatbot.",
				r'thank	you|thanks':	"You're	welcome!",
				r'bye|goodbye':	"Goodbye!	Have	a	nice	day!",
				r'(help|support)':	"Sure!	How	can	I	assist	you?",
 }
```

```python
def	get_response(user_input):
				user_input	=	user_input.lower()
				for	pattern,	response	in	patterns.items():
								if	re.search(pattern,	user_input):
												return	response
				return	"Sorry,	I	didn't	understand	that.	Can	you	try	rephrasing?"
```

```python
def chatbot():
    print("Chatbot:	Hi!")
    while	True:
        user_input	=	input("You:	")
        if	user_input.lower()	in	['bye',	'goodbye']:
            print("Chatbot:	Goodbye!	Have	a	nice	day!")
            break
        response	=	get_response(user_input)
        print(f"Chatbot:	{response}")
```

```python
if __name__	=="__main__":
			chatbot()
```
