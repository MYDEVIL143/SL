```python
import spacy

nlp = spacy.load("en_core_web_sm")

text = """On January 20, 2025, the American company SpaceX launched its latest Falcon 9 rocket 
from Cape Canaveral, Florida. The CEO, Elon Musk, said the mission will cost about $500 million 
and last two years."""

doc = nlp(text)

print("Named Entities:\n")
for ent in doc.ents:
    print(ent.text, "->", ent.label_)
```
