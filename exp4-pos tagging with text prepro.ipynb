```python
import re
import string
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer

# Download required resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Input text
text = """Artificial Intelligence is transforming industries around the world.
It is being used in healthcare, finance, education and many other fields.
The technology allows machines to learn, adapt and perform tasks that previously required human intelligence."""

# Preprocessing
text = re.sub(r'[^\w\s]', '', text.lower())
tokens = word_tokenize(text)

# Remove stopwords and punctuation
stop_words = set(stopwords.words('english'))
tokens_clean = [word for word in tokens if word not in stop_words and word not in string.punctuation]

# Stemming
ps = PorterStemmer()
stemmed = [ps.stem(word) for word in tokens_clean]
print("Stemmed Words:\n", stemmed)

# Lemmatization
wnl = WordNetLemmatizer()
lemmatized = [wnl.lemmatize(word) for word in tokens_clean]
print("\nLemmatized Words:\n", lemmatized)
```
