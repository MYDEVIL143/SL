```python
import re

# Input text
text = """Contact me at john.doe@example.com or support@company.org for assistance.
For inquiries, call 123-456-7890 or 987.654.3210. 
You can also reach me at (555) 555-1234 and 7001093077."""

# Regex patterns
email_pattern = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,7}\b'
phone_pattern = r'\b(?:\d{3}[-.\s]?)?\d{3}[-.\s]?\d{4}\b|\(\d{3}\)\s?\d{3}[-.\s]?\d{4}\b|\b\d{10}\b'

# Find matches
emails = re.findall(email_pattern, text)
phones = re.findall(phone_pattern, text)

# Output results
print(f"Emails found: {emails}")
print(f"Phone numbers found: {phones}")
```
